```python
from typing import List, Dict

def process_audio_for_training(
    audio_folder: str,
    output_dir: str,
    max_files: int = None
) -> List[Dict]:
    """
    Xử lý toàn bộ file âm thanh trong thư mục để tạo dữ liệu train mô hình.
    
    Args:
        audio_folder (str): Đường dẫn đến thư mục chứa file âm thanh.
        output_dir (str): Thư mục lưu đặc trưng MFCC.
        max_files (int, optional): Số lượng file tối đa để xử lý (None = tất cả).
    
    Returns:
        List[Dict]: Danh sách chứa thông tin mỗi file: 
                    {'file_path': str, 'mfcc': np.ndarray, 'sample_rate': int}
    """
    # Hàm xử lý âm thanh
    def process_audio(audio_data: np.ndarray, sample_rate: int) -> np.ndarray:
        # Chuẩn hóa biên độ
        audio_data = audio_data / np.max(np.abs(audio_data)) if np.max(np.abs(audio_data)) != 0 else audio_data
        # Cắt khoảng lặng
        audio_data, _ = librosa.effects.trim(audio_data, top_db=20)
        return audio_data

    # Tạo thư mục đầu ra
    os.makedirs(output_dir, exist_ok=True)

    results = []
    file_count = 0

    try:
        for root, _, files in os.walk(audio_folder):
            for file in files:
                if max_files and file_count >= max_files:
                    break
                if file.endswith('.wav'):  # Chỉ xử lý file WAV
                    audio_file_path = os.path.join(root, file)
                    try:
                        # Đọc âm thanh
                        audio_data, sample_rate = librosa.load(audio_file_path, sr=16000)  # Resample về 16kHz
                        
                        # Xử lý âm thanh
                        processed_audio = process_audio(audio_data, sample_rate)
                        
                        # Trích xuất MFCC
                        mfcc = librosa.feature.mfcc(y=processed_audio, sr=sample_rate, n_mfcc=13)
                        
                        # Lưu MFCC vào file
                        file_id = os.path.splitext(file)[0]
                        output_path = os.path.join(output_dir, f"{file_id}_mfcc.npy")
                        np.save(output_path, mfcc)
                        
                        # Lưu thông tin
                        results.append({
                            'file_path': audio_file_path,
                            'mfcc': mfcc,
                            'sample_rate': sample_rate
                        })
                        
                        file_count += 1
                    except Exception as e:
                        print(f"Lỗi khi xử lý file {audio_file_path}: {e}")
            if max_files and file_count >= max_files:
                break
    except FileNotFoundError:
        print(f"Lỗi: Thư mục '{audio_folder}' không tồn tại.")
        return []
    
    return results

# Sử dụng hàm
if __name__ == "__main__":
    audio_folder = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\train\waves"
    output_dir = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\train\mfcc_features"

    # Xử lý toàn bộ folder
    results = process_audio_for_training(
        audio_folder=audio_folder,
        output_dir=output_dir,
        max_files=None  # Xử lý tất cả file
    )
    
    # In số lượng file đã xử lý
    print(f"Đã xử lý {len(results)} file âm thanh. Đặc trưng MFCC được lưu tại {output_dir}")
```

# Model conformer

```python
import os
import torch
import torchaudio
import librosa
import numpy as np
from torch import nn
from torch.utils.data import Dataset, DataLoader
from jiwer import wer, cer
from typing import List, Dict, Tuple
from conformer import Conformer

# Đường dẫn dữ liệu
TRAIN_AUDIO_PATH = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\train\waves"
TRAIN_TEXT_PATH = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\train\prompts.txt"
TEST_AUDIO_PATH = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\test\waves"
TEST_TEXT_PATH = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\test\prompts.txt"
OUTPUT_DIR = r"D:\Nam_2\CDIO 3\Speech_to_Text\vivos\model"

# Tạo bộ ký tự (character vocabulary)
def create_vocab(texts: List[str]) -> Dict[str, int]:
    chars = sorted(set(''.join(text.lower() for text in texts)))
    char_to_idx = {char: idx + 1 for idx, char in enumerate(chars)}  # 0 for blank (CTC)
    char_to_idx['<blank>'] = 0
    return char_to_idx

# Đọc prompts.txt
def read_prompts(file_path: str) -> Dict[str, str]:
    prompts = {}
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) > 1:
                    prompts[parts[0]] = ' '.join(parts[1:]).lower()
    except Exception as e:
        print(f"Lỗi khi đọc file {file_path}: {e}")
    return prompts

# Dataset cho VIVOS
class VIVOSDataset(Dataset):
    def __init__(self, audio_path: str, text_path: str, char_to_idx: Dict[str, int]):
        self.audio_files = []
        self.texts = []
        self.prompts = read_prompts(text_path)
        self.char_to_idx = char_to_idx
        
        for root, _, files in os.walk(audio_path):
            for file in files:
                if file.endswith('.wav'):
                    file_id = os.path.splitext(file)[0]
                    if file_id in self.prompts:
                        self.audio_files.append(os.path.join(root, file))
                        self.texts.append(self.prompts[file_id])
        
    def __len__(self):
        return len(self.audio_files)
    
    def __getitem__(self, idx):
        # Đọc âm thanh
        waveform, sample_rate = torchaudio.load(self.audio_files[idx])
        if sample_rate != 16000:
            waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)(waveform)
        waveform = waveform.squeeze(0)  # Chuyển thành 1D
        
        # Tiền xử lý âm thanh
        waveform = waveform / torch.max(torch.abs(waveform))  # Chuẩn hóa
        waveform = waveform.numpy()
        waveform, _ = librosa.effects.trim(waveform, top_db=20)  # Cắt khoảng lặng
        
        # Trích xuất MFCC
        mfcc = librosa.feature.mfcc(y=waveform, sr=16000, n_mfcc=13)
        mfcc = torch.tensor(mfcc, dtype=torch.float32)  # [n_mfcc, time]
        
        # Chuyển text thành chuỗi số
        text = self.texts[idx]
        text_ids = [self.char_to_idx[char] for char in text if char in self.char_to_idx]
        text_ids = torch.tensor(text_ids, dtype=torch.int64)
        
        return mfcc, text_ids, len(mfcc[0]), len(text_ids)

# Collate function cho DataLoader
def collate_fn(batch):
    mfccs, text_ids, mfcc_lengths, text_lengths = zip(*batch)
    mfccs = nn.utils.rnn.pad_sequence(mfccs, batch_first=True)  # [batch, n_mfcc, max_time]
    text_ids = nn.utils.rnn.pad_sequence(text_ids, batch_first=True, padding_value=0)  # [batch, max_len]
    return mfccs, text_ids, torch.tensor(mfcc_lengths), torch.tensor(text_lengths)

# Mô hình Conformer-CTC
class ConformerCTC(nn.Module):
    def __init__(self, input_dim: int, num_classes: int, num_layers: int = 4):
        super().__init__()
        self.conformer = Conformer(
            dim=input_dim,
            depth=num_layers,
            dim_head=64,
            heads=8,
            ff_mult=4,
            conv_expansion_factor=2,
            conv_kernel_size=31,
            attn_dropout=0.1,
            ff_dropout=0.1,
            conv_dropout=0.1
        )
        self.fc = nn.Linear(input_dim, num_classes)
    
    def forward(self, x, lengths):
        x = x.transpose(1, 2)  # [batch, time, n_mfcc]
        x = self.conformer(x)  # [batch, time, dim]
        x = self.fc(x)  # [batch, time, num_classes]
        return x

# Hàm huấn luyện một epoch
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    for mfccs, text_ids, mfcc_lengths, text_lengths in dataloader:
        mfccs, text_ids = mfccs.to(device), text_ids.to(device)
        mfcc_lengths, text_lengths = mfcc_lengths.to(device), text_lengths.to(device)
        
        optimizer.zero_grad()
        outputs = model(mfccs, mfcc_lengths).log_softmax(2)  # [batch, time, num_classes]
        
        loss = criterion(outputs.transpose(0, 1), text_ids, mfcc_lengths, text_lengths)  # CTC Loss
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    return total_loss / len(dataloader)

# Hàm đánh giá
def evaluate(model, dataloader, char_to_idx, device):
    model.eval()
    idx_to_char = {idx: char for char, idx in char_to_idx.items()}
    hypotheses, references = [], []
    
    with torch.no_grad():
        for mfccs, text_ids, mfcc_lengths, text_lengths in dataloader:
            mfccs = mfccs.to(device)
            outputs = model(mfccs, mfcc_lengths).log_softmax(2)  # [batch, time, num_classes]
            
            # Giải mã CTC
            for i in range(outputs.size(0)):
                output = outputs[i, :mfcc_lengths[i]].argmax(1).cpu().numpy()
                hyp = []
                for idx in output:
                    if idx != 0 and (not hyp or hyp[-1] != idx):  # Bỏ blank và lặp
                        hyp.append(idx)
                hyp_text = ''.join(idx_to_char.get(idx, '') for idx in hyp if idx in idx_to_char)
                ref_text = ''.join(idx_to_char[idx.item()] for idx in text_ids[i] if idx.item() in idx_to_char)
                
                hypotheses.append(hyp_text)
                references.append(ref_text)
    
    wer_score = wer(references, hypotheses)
    cer_score = cer(references, hypotheses)
    
    return wer_score, cer_score, hypotheses[:5], references[:5]

# Main
def main():
    # Tạo thư mục đầu ra
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    
    # Đọc dữ liệu
    train_prompts = read_prompts(TRAIN_TEXT_PATH)
    test_prompts = read_prompts(TEST_TEXT_PATH)
    all_texts = list(train_prompts.values()) + list(test_prompts.values())
    char_to_idx = create_vocab(all_texts)
    
    # Tạo dataset và dataloader
    train_dataset = VIVOSDataset(TRAIN_AUDIO_PATH, TRAIN_TEXT_PATH, char_to_idx)
    test_dataset = VIVOSDataset(TEST_AUDIO_PATH, TEST_TEXT_PATH, char_to_idx)
    train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=collate_fn, num_workers=4)
    test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, collate_fn=collate_fn, num_workers=4)
    
    # Khởi tạo mô hình
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = ConformerCTC(input_dim=13, num_classes=len(char_to_idx)).to(device)
    criterion = nn.CTCLoss(blank=0, zero_infinity=True)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    # Huấn luyện
    num_epochs = 20
    best_wer = float('inf')
    for epoch in range(num_epochs):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
        wer_score, cer_score, hyp_samples, ref_samples = evaluate(model, test_loader, char_to_idx, device)
        
        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, WER: {wer_score:.4f}, CER: {cer_score:.4f}")
        for i in range(min(5, len(hyp_samples))):
            print(f"Sample {i+1} - Hyp: {hyp_samples[i]} | Ref: {ref_samples[i]}")
        
        # Lưu mô hình nếu WER tốt hơn
        if wer_score < best_wer:
            best_wer = wer_score
            torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, 'best_model.pt'))
            print(f"Đã lưu mô hình tốt nhất tại {os.path.join(OUTPUT_DIR, 'best_model.pt')}")
    
    # Đánh giá cuối cùng
    print("\n=== Kết quả đánh giá cuối cùng ===")
    print(f"Word Error Rate (WER): {wer_score:.4f}")
    print(f"Character Error Rate (CER): {cer_score:.4f}")
    print("\nMẫu dự đoán và tham chiếu (5 mẫu đầu tiên):")
    for i in range(len(hyp_samples)):
        print(f"Mẫu {i+1}:")
        print(f"  Dự đoán: {hyp_samples[i]}")
        print(f"  Tham chiếu: {ref_samples[i]}")
        print()

if __name__ == "__main__":
    main()
```
